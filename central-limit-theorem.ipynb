```python
#Libraries
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

#### Randomly select a series with 5000 observations between 0 and 500. This data will be our population. 

```python
population = np.random.randint(501,size=5000)
```

```python
#Histogram of randomly selected series
sns.distplot(population)
plt.xlabel('Observation')
plt.show()
```

```python
print(f"Mean of the population: {population.mean()}")
```

#### Select randomly 50 samples each time and take their average. means_of_samples will be our list which consist of samples averages.

```python
means_of_samples=[]
for i in range(1,1000):
    means_of_samples.append(np.random.choice(population,size=50).mean())
```

```python
sns.distplot(means_of_samples)
sns.distplot(population)
plt.xlabel('Mean of samples and population data')
plt.plot();
```

## Conclusion

#### Central limit theory states that the mean of randomly selected samples from a series (the series does not need to be normally distributed) will be normally distributed around the mean of the series (the size of the sample (n) must be greater than 30). As the number of selected samples increases (n), it converges the normal distribution.
